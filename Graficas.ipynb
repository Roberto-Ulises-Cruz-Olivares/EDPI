```python
import matplotlib.pyplot as plt
import numpy as np

def function(x,y):

    plt.style.use('ggplot')
    fig, ax = plt.subplots(figsize=(10, 6))  # Ajusta el tamaño para que sea más ancho que alto

    ax.plot(x, y, '-*', label='Función')  # Línea de la función
    ax.set_title('Primer ejercicio EDPI')
    ax.set_xlabel('Eje x')
    ax.set_ylabel('Eje y')

    # Ajustar los límites de los ejes para una mejor visualización

    ax.set_xlim(x[0] - 1, x[-1] + 1)  # Agregar un margen en el eje x
    ax.set_ylim(0, max(y) + 10)  # Ajustar el límite superior del eje y

    # Añadir una leyenda para el gráfico
    ax.legend()

    plt.tight_layout()  # Ajustar el diseño del gráfico
    plt.show()  # Mostrar el gráfico

#DATOS
x = np.linspace(-4, 4, 100)
y=x**2

Grafica=function(x,y)
print(f"Grafica: {Grafica}")
```
