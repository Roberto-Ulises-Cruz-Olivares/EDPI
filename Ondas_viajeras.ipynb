# Funcion original:
$$
f(x´) = e^{-\frac{(x´-0.5)^2}{0.01} }
$$

Como primer paso Grafiquemos f(x) para ver como es :

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funcion
def f(x_prima):
    return np.exp(-((x_prima - 0.5)**2) / 0.01)

#Graficamos la funcion

x = np.linspace(0, 1, 1000)
f_original = f(x)


# Graficar todo
plt.figure(figsize=(20, 20))

plt.subplot(4, 1, 1)
plt.plot(x, f_original, label="Función original $f(x')$", color='red')
plt.title("Función original $f(x')$")
plt.grid()
plt.legend()
```

# $a)$ Mostrar que al hacer el cambio de variable $x´ = x -ct$ la pertubacion se desplaza a la derecha

```python
c = 0.2
t = 1.2
desplazamiento_derecha = x - c*t
f_cambioderecha = f( desplazamiento_derecha )

#Grafiquemos la funcion hacia la derecha
plt.figure(figsize=(20, 20))
f_derecha = f_cambioderecha

plt.subplot(4, 1, 2)
plt.plot(x, f_derecha, label="Grafica hacia la derecha", color='red')
plt.title("a) Perturbación hacia la derecha (x' = x - ct)")
plt.grid()
plt.legend()
```

# $b)$ Mostrar que al hacer el cambio de variable $x´ = x + ct$ la pertubacion se desplaza a la izquierda

```python
c = 0.2
t = 1.2
desplazamiento_izq = x + c*t
f_cambioizq = f( desplazamiento_izq )

#Grafiquemos la funcion hacia la izquierda
plt.figure(figsize=(20, 20))
f_izq = f_cambioizq

plt.subplot(4, 1, 2)
plt.plot(x, f_izq, label="Grafica hacia la derecha", color='green')
plt.title("a) Perturbación hacia la derecha (x' = x - ct)")
plt.grid()
plt.legend()
```


# $ c ) $ Grafica la superposicion de las ondas viajeras de los incisos $ a) $ y $ b) $


```python
# Superposicion graficas juntas
f_total = f_derecha + f_izq

# Grafica de superposicion
plt.figure(figsize=(20, 20))
plt.subplot(4, 1, 4)
plt.plot(x, f_total, label="Superposición", color='blue')
plt.title("c) Superposición de las dos perturbaciones")
plt.grid()
plt.legend()
```

# $d)$ Vamos observar ahora como nuestro tiempo varia conforme a la onda

```python
tiempo =  [0.1 , 1, 1.2, 1.5, 2] # hacemos una lista con los tiempos que queremos que tome

plt.figure(figsize=(10, 10))

for t in tiempo: # que grafique los tiempos que queremos en una grafica

    desplazamiento_derecha = x - c*t
    desplazamiento_izq = x + c*t


    f_derecha = f( desplazamiento_derecha ) # definimos la de la derecha
    f_izq = f( desplazamiento_izq ) # definimos la de la izq

    f_total = f_derecha + f_izq # definimos la que tiene superposicion

    plt.plot(x, f_total, label=f't = {t}') # ploteamos todas con nuestros distintos t

plt.title("Onda a traves del tiempo")
plt.grid()
plt.legend()
plt.show()
```
