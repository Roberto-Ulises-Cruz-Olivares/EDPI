# **EJEMPLO**

```python
from sympy import fourier_series, pi, plot, Piecewise, exp
from sympy.abc import x
```

```python
f=Piecewise(
    (0, x < -2),
    (x+2,(x >= -2) & (x < 0)),
    (0,(x<=2)))
```

```python
plot(f, (x,-2,2),line_color ='red')
```

```python
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

# ***Problema 1 FUNCION SENO***

```python
f1=Piecewise(
    (-exp(2*x), (x >= -1) & (x < 0)),
    ((exp(2*x), x >= 0) ) )
```

```python
plot(f1, (x,-1,1),line_color ='red')
```

```python
s_0 = fourier_series(f1, (x, -1, 1))
s_1 = s_0.truncate(n = 3)
s_2 = s_0.truncate(n = 5)
s_3 = s_0.truncate(n = 100)
p_0 = plot(f1, s_1, s_2, s_3, (x, -1, 1), show=False, legend=True)
```

```python
s_0
```

```python
p_0[0].label = 'x'
p_0[1].label = 'n=3'
p_0[2].label = 'n=5'
p_0[3].label = 'n=100'

p_0.show()
```
