# ***Ejercicio de Clase Plot***

***Importamos las librerias de sympy para poder plotear en 3D y la segunda para pdoer definir los simbolos para definir nuestras variables***

```python
from sympy.plotting import plot3d #Libreria
from sympy import *
```

***Definimos nuestras variables***

```python
x,y=symbols('x y') #asigna las letras de las variables a usar
```

***Ploteamos en 3D con nuestra funcion plot3d() en un intervalo para x y y***

```python
plot3d( x**3-x*y + y**3, (x,-5,5),(y,-5,5))
```
